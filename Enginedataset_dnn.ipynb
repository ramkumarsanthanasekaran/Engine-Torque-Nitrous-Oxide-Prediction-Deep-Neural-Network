```python
input = pd.read_excel("inputs_Engine.xlsx", index_col=False)
output = pd.read_excel("output_Engine.xlsx", index_col=False)

# Transpose DataFrames
input = input.T
output = output.T

# Reset index to default integer index
input.reset_index(inplace=True)
output.reset_index(inplace=True)
```

```python
input['Fuel_Rate'] = pd.to_numeric(input['Fuel_Rate'], errors='coerce')
```

```python
output['Torque'] = pd.to_numeric(output['Torque'], errors='coerce')
```

```python
output["Nitrous_oxide_emissions"]=output["Nitrous_oxide_emissions"].astype(np.float64)
```

```python
combined=pd.concat([input,output],axis=1)
combined
```

```python
combined.dropna(axis=0,inplace=True)
combined
```

```python
input1=combined.pop("Fuel_Rate")
```

```python
input2=combined.pop("Speed")
```

```python
input1
```

```python
input2
```

```python
input=pd.concat([input1,input2],axis=1)
```

```python
output_new=combined
```

```python
output_new
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
inp = sc.fit_transform(input)

sc = StandardScaler()
norm_out = sc.fit_transform(output_new)
```

```python
model=tf.keras.models.Sequential([
    tf.keras.layers.Dense(20, input_shape=(2,), activation='relu'),
    tf.keras.layers.Dense(32,activation='relu'),
    tf.keras.layers.Dense(2,activation='linear'),
])
```

```python
model.compile(loss="mse",optimizer='sgd',metrics=['accuracy'])
```

```python
model.fit(inp,norm_out,epochs=500)
```

```python
y=model.predict(inp[56:67])
sc.inverse_transform(y)
```
